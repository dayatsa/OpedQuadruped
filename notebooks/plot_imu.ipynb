```python
import matplotlib.pyplot as plt
```

```python
true_value = [-40,-30,-25,-20,-15,-10,-5,0,5,10,15,20,25,30,40]
predict_roll = [-39.16, -30.13, -24.86, -20.26, -14.77, -9.83, -5.13, 0.24, 5.15, 9.87, 15.17, 19.89, 25.22, 30.44, 39.90]
predict_pitch = [-40.37, -29.88, -25.42, -20.45, -14.82, -10.67, -5.28, 0.36, 4.81, 9.71, 14.90, 19.74, 24.74, 29.53, 39.58]
```

```python
predicted_value = predict_roll

plt.figure()
plt.scatter(true_value, predicted_value, c='crimson')

p1 = max(max(predicted_value), max(true_value))
p2 = min(min(predicted_value), min(true_value))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.grid()
plt.show()
```
