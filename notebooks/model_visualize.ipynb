```python
import numpy as np
import random
```

```python
def printTable(q_table):
  data = ""
  for state in q_table:
    data = ""
    for i in range(len(state)):
      if (i == 30 or i == 31):
        data = data + "| "
      data = data + str(np.argmax(state[i])) + " "
    print(data)
```

```python
path_x = '/content/OpedQuadruped/oped/oped_teleopp/model/x/model_x_trained.npy'
path_y = '/content/OpedQuadruped/oped/oped_teleopp/model/y/model_y_trained.npy'
```

```python
q_table_x = np.load(path_x)
printTable(q_table_x)
```
